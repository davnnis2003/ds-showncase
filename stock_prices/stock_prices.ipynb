```python
import requests
import json
import pandas as pd
```

```python
import yaml
from yaml.loader import SafeLoader

with open('keys.yaml') as f:
    keys = yaml.load(f, Loader=SafeLoader)
    print(keys)
```

```python
url = "https://apidojo-yahoo-finance-v1.p.rapidapi.com/stock/v3/get-historical-data"

key = keys['key']
host = keys['host']

symbol = '0005.HK'
region = 'HK'
```

```python
querystring = {"symbol":symbol,"region":region}

headers = {
    'x-rapidapi-key': key,
    'x-rapidapi-host': host
    }

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```

```python
import json

messages = json.loads(response.text)
print(messages['prices'][0])
```

```python
df = pd.json_normalize(messages)
df.head()
```

```python
pd.json_normalize(data = df['prices'],
                  record_path = 'date',
                  meta = ['open','high', 'low', 'close', 'volume','adjclose']
                 )
```
